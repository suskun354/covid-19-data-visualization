```python
df =pd.read_csv("corona_dat.csv")
```

```python
import datetime
df['date']=pd.to_datetime(df['date'])
df['date'].dtype
```

```python
df.info()
```

```python
df_yarisan_grf=df[['date','Turkey','Italy','Spain','Germany','China','Russia','US']]
```

```python
df_yarisan_grf.head()
```

```python
#index Tarih Sayısına göre girilmiş ama her zaman bu şekilde olmayabilir
#Tarih veri türü çevrilir ve tarihe göre sıralama (index)
df_yarisan_grf.set_index('date',inplace=True)
df_yarisan_grf.head()
```

```python
import seaborn as sns
#Günlük vaka sayısı
sns.lineplot(data=df_yarisan_grf);
```

```python
df_yarisan_grf=df_yarisan_grf.cumsum(axis=0)
df_yarisan_grf
```

```python
plt.figure(figsize=(12,6));
sns.lineplot(data=df_yarisan_grf)
plt.xticks(rotation=90);
plt.title("Günlük vaka sayısı");
plt.show()
```
